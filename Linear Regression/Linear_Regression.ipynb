```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
dataset = pd.read_csv('student-mat.csv', sep = ";")
```

```python
#converting categorical data to dummies
school = pd.get_dummies(dataset['school'], drop_first=True)
sex = pd.get_dummies(dataset['sex'], drop_first=True)
address  = pd.get_dummies(dataset['address'], drop_first=True)
famsize = pd.get_dummies(dataset['famsize'], drop_first=True)
Pstatus  = pd.get_dummies(dataset['Pstatus'], drop_first=True)
Mjob  = pd.get_dummies(dataset['Mjob'], drop_first=True)
Fjob = pd.get_dummies(dataset['Fjob'], drop_first=True)
reason  = pd.get_dummies(dataset['reason'], drop_first=True)
guardian = pd.get_dummies(dataset['guardian'], drop_first=True)
schoolsup = pd.get_dummies(dataset['schoolsup'], drop_first=True)
famsup = pd.get_dummies(dataset['famsup'], drop_first=True)
paid = pd.get_dummies(dataset['paid'], drop_first=True)
activities  = pd.get_dummies(dataset['activities'], drop_first=True)
nursery  = pd.get_dummies(dataset['nursery'], drop_first=True)
higher = pd.get_dummies(dataset['higher'], drop_first=True)
internet = pd.get_dummies(dataset['internet'], drop_first=True)
romantic = pd.get_dummies(dataset['romantic'], drop_first=True)
```

```python
dataset = dataset.drop(labels = ['school', 'sex', 'address', 'famsize', 'Pstatus', 'Mjob', 'Fjob','reason', 'guardian', 'schoolsup','famsup','paid', 'activities', 'nursery', 'higher', 'internet', 'romantic'], axis = 1)
```

```python
dataset = pd.concat( [dataset,school, sex, address, famsize, Pstatus, Mjob, Fjob,reason, guardian, schoolsup,famsup,paid, activities, nursery, higher, higher, internet, romantic], axis=1)
```

```python
dataset.head()
```

```python
dataset.dtypes
```

```python
y1 = dataset['G1']
y2 = dataset['G2']
y3 = dataset['G3']
```

```python
X = dataset.drop(labels = ['G1', 'G2', 'G3'], axis = 1)
```

```python
X2 = X
X3 = X
```

```python
from sklearn.model_selection import train_test_split 
X_train1, X_test1, y_train1, y_test1 = train_test_split(X, y1, test_size=0.2, random_state=0)
```

```python
from sklearn.model_selection import train_test_split 
X_train2, X_test2, y_train2, y_test2 = train_test_split(X2, y2, test_size=0.2, random_state=0)
```

```python
from sklearn.model_selection import train_test_split 
X_train3, X_test3, y_train3, y_test3 = train_test_split(X3, y3, test_size=0.2, random_state=0)
```

```python
from sklearn.linear_model import LinearRegression
```

```python
regressor1 = LinearRegression()  
regressor2 = LinearRegression()  
regressor3 = LinearRegression()  
regressor1.fit(X_train1, y_train1)
regressor2.fit(X_train2, y_train2)
regressor3.fit(X_train3, y_train3)
```

```python
#To retrieve the intercept:
print(regressor1.intercept_)
#For retrieving the slope:
print(regressor1.coef_)
```

```python
print(regressor2.intercept_)
print(regressor2.coef_)
```

```python
print(regressor3.intercept_)
print(regressor3.coef_)
```

```python
y_pred1 = regressor1.predict(X_test1)
y_pred2 = regressor2.predict(X_test2)
y_pred3 = regressor3.predict(X_test3)
```

```python
from sklearn import metrics
```

```python
print('Mean Absolute Error:', metrics.mean_absolute_error(y_test1, y_pred1))  
print('Mean Squared Error:', metrics.mean_squared_error(y_test1, y_pred1))  
print('Root Mean Squared Error:', np.sqrt(metrics.mean_squared_error(y_test1, y_pred1)))
```

```python
print('Mean Absolute Error:', metrics.mean_absolute_error(y_test2, y_pred2))  
print('Mean Squared Error:', metrics.mean_squared_error(y_test2, y_pred2))  
print('Root Mean Squared Error:', np.sqrt(metrics.mean_squared_error(y_test2, y_pred2)))
```

```python
print('Mean Absolute Error:', metrics.mean_absolute_error(y_test3, y_pred3))  
print('Mean Squared Error:', metrics.mean_squared_error(y_test3, y_pred3))  
print('Root Mean Squared Error:', np.sqrt(metrics.mean_squared_error(y_test3, y_pred3)))
```
